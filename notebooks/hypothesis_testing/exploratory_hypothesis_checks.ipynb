```python
import os
import pandas as pd

# detect project root (the one that contains 'data' folder)
cwd = os.getcwd()
print("Current working directory:", cwd)

# try to find project root dynamically
while True:
    if "data" in os.listdir(cwd):
        project_root = cwd
        break
    new = os.path.dirname(cwd)
    if new == cwd:  # reached disk root
        raise FileNotFoundError("Could not locate the project root containing the 'data' folder.")
    cwd = new

print("Detected project root:", project_root)

# build absolute path
data_path = os.path.join(project_root, "data", "processed", "insurance_data_cleaned.csv")
print("Loading:", data_path)

df = pd.read_csv(data_path)
df.head()
```
